```python
import pandas as pd #importar a biblioteca panda como pd (apelido)
```

```python
df.columns #nome das colunas
```

```python
renomear_colunas = {
    'work_year': 'ano_trabalho',
    'experience_level': 'nivel_experiencia',
    'employment_type': 'tipo_emprego',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda_salario',
    'salary_in_usd': 'salario_em_usd',
    'employee_residence': 'residencia_empregado',
    'remote_ratio': 'taxa_remoto',
    'company_location': 'localizacao_empresa',
    'company_size': 'tamanho_empresa'
} #dicionario de tradução dos nomes das colunas
df.rename(columns=renomear_colunas, inplace=True) #renomear colunas
print(df.columns)
```

```python
df["nivel_experiencia"].value_counts() #valores das colunas
```

```python
df["tipo_emprego"].value_counts()
```

```python
df["taxa_remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
nivel_experiencia_traducao = {
    'SE': 'Nível Sênior',
    'MI': 'Nível Pleno',
    'EN': 'Nível Júnior',
    'EX': 'Nível Executivo'
} #dicionario de tradução das categorias da coluna de nivel de experiencia

df['nivel_experiencia'] = df['nivel_experiencia'].replace(nivel_experiencia_traducao) #renomear categorias
print(df['nivel_experiencia'].value_counts())
```

```python
tipo_emprego_traducao = {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
}#dicionario de tradução das categorias da coluna de tipo de emprego

df['tipo_emprego'] = df['tipo_emprego'].replace(tipo_emprego_traducao)
print(df['tipo_emprego'].value_counts())
```

```python
tamanho_empresa_traducao = {
    "S": "pequena",
    "M": "Média",
    "L": "Grande"
} #dicionario de tradução das categorias da coluna tamanho empresa
df["tamanho_empresa"] = df["tamanho_empresa"].replace(tamanho_empresa_traducao)
print(df["tamanho_empresa"].value_counts())
```

```python
taxa_remoto_traducao = {
    0: 'Presencial',
    50: 'Hibrido',
    100: 'Remoto'
} #dicionario de tradução das categorias da coluna de taxa de trabalho remoto
df['taxa_remoto'] = df['taxa_remoto'].replace(taxa_remoto_traducao)
print(df['taxa_remoto'].value_counts())
```

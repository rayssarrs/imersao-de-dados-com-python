```python
import pandas as pd
import numpy as np #biblioteca para manipulação de dados numéricos

df_salarios = pd.DataFrame({ #criar um dataframe do zero
     "nome": ["Astrid", "Isa", "Israela", "Mel", "Rayssa"], #criação de colunas, lista sempre entre []
     "salario": [np.nan, 16000, 18000, np.nan, 20000] #np.nan - valor nulo (vai preencher com média do salário). não usa "" para número
 })
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2)) #df_salarios - base. "salario_media" - nova coluna. fillna - preencher valores nulos. mean - calcula média. round(2) - arredondamento de duas casas decimais
df_salarios #mostra dataframe
```

```python
df_temperaturas = pd.DataFrame({
    "Dia da semana": ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_ffill"] = df_temperaturas["Temperatura"].ffill() #ffill (forward fill) - completa com valor interior
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    "Dia da semana": ["Segunda-feira", "Terça-feira", "Quarta-feira", "Quinta-feira", "Sexta-feira"],
    "Temperatura": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["preenchido_bfill"] = df_temperaturas["Temperatura"].bfill() #bfill (back fill) - completa com valor posterior
df_temperaturas
```
